```python
import numpy as np
import pandas as pd
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
```

```python
movies = movies[['id','title','genres','overview','keywords']]
credits = credits[['title','cast','crew']]
```

```python
movies.dropna(inplace=True)
credits.isnull().sum()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.iloc[0].genres
```

```python
import ast
temp1=[]
for i in movies.genres:
    temp2=[]
    for j in ast.literal_eval(i):
        temp2.append(j['name'])
    temp1.append(temp2)
movies.genres = temp1
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
temp1=[]
for i in movies.keywords:
    temp2=[]
    for j in ast.literal_eval(i):
        temp2.append(j['name'])
    temp1.append(temp2)
movies.keywords = temp1
movies.head()
```

```python
temp1=[]
for i in movies.cast:
    temp2=[]
    c=0
    for j in ast.literal_eval(i):
        if c<3:
            temp2.append(j['name'])
            c+=1
        else:
            break
    temp1.append(temp2)
movies.cast=temp1
movies.head()
```

```python
movies.iloc[0].crew
```

```python
temp1=[]
for i in movies.crew:
    temp2=[]
    c=0
    for j in ast.literal_eval(i):
        if j['job'] == 'Director':
            temp2.append(j['name'])
            break
    temp1.append(temp2)
movies.crew=temp1
movies.head()
```

```python
movies.overview = movies.overview.apply(lambda x : x.split(" "))
movies.head()
```

```python
movies.genres = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies.overview = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])

movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
df = movies[['title' ,'id','tags']]
df
```

```python
df['tags'] =  df['tags'].apply(lambda x :" ".join(x))
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
t2=[]
for i in df['tags']:
    t1 = []
    for j in i.split():
        t1.append(ps.stem(j))
    t2.append(" ".join(t1))
df['tags']=t2
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
print(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommeded(movie):
    movie_ind=df[df['title']==movie].index[0]
    dist = similarity[movie_ind]
    movie_list = sorted(list(enumerate(dist)),reverse=True,key=lambda x : x[1])[1:6]

    for i in movie_list:
        print(df.iloc[i[0]].title)

```

```python
recommeded('Avatar')
```
